```python
from io import StringIO
import pandas as pd

data = StringIO(r"""
Method	Per Call (ms)	% Total Time	Avg. Allocated (MiB)	% Total Mem	nV	Qubits	Hdim
Pool gradient element	7.45E-03	0.5	6.42E-03	7	1	2	4
Parameter gradient element	1.70E-02	2	1.11E-02	21.1	1	2	4
Pool gradient element	1.07E-01	6.6	2.55E-02	6.1	2	4	16
Parameter gradient element	1.50E-01	33.1	7.84E-02	66.8	2	4	16
Pool gradient element	6.30E-01	0.7	0.309	1.1	3	6	64
Parameter gradient element	3.98	96.9	1.08	86.1	3	6	64
""")

df = pd.read_csv(data, delimiter="\t")
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_palette("hls")
sns.set_style("ticks")
```

```python
sns.lineplot(data=df, x="Hdim", y="Per Call (ms)", hue="Method", marker='o')
plt.xlabel("Hilbert Dimension")
plt.xscale('log', base=2)
# plt.yscale('log')
plt.show()
```

```python
sns.lineplot(data=df, x="Hdim", y="Avg. Allocated (MiB)", hue="Method", marker='o')
plt.xlabel("Hilbert Dimension")
plt.xscale('log', base=2)
# plt.yscale('log')
plt.show()
```
